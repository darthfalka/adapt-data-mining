### Summary statistics questions:
- [Level 1 included] Whats the main criteria by which this table breaks down the population? Indigenous vs Non Indigenous
- [Level 2 included] Do we have breakdown by gender? Yes
- [Level 3 not included] Do we have breakdown by age? Yes

```python
# Level 1
df1['Tot_Indigenous_P_by_Tot_Tot_P'] = df1['Tot_Indigenous_P']/df1['Tot_Tot_P'] * 100 # total indigenous by total people
df1['Tot_Non_Indigenous_P_by_Tot_Tot_P'] = df1['Tot_Non_Indigenous_P']/df1['Tot_Tot_P'] * 100 # total non-indigenous by total people
df1_level1 = ['Tot_Indigenous_P_by_Tot_Tot_P', 'Tot_Non_Indigenous_P_by_Tot_Tot_P']

# Level 2
# Male
df1['Tot_Indigenous_M_by_Tot_Tot_M'] = df1['Tot_Indigenous_M']/df1['Tot_Tot_M'] * 100 # male indigenous by total male
df1['Tot_Non_Indigenous_M_by_Tot_Tot_M'] = df1['Tot_Non_Indigenous_M']/df1['Tot_Tot_M'] * 100 # male non-indigenous by total male

# Female
df1['Tot_Indigenous_F_by_Tot_Tot_F'] = df1['Tot_Indigenous_F']/df1['Tot_Tot_F'] * 100 # female indigenous by total female
df1['Tot_Non_Indigenous_F_by_Tot_Tot_F'] = df1['Tot_Non_Indigenous_F']/df1['Tot_Tot_F'] * 100 # female non-indigenous by total female
```

### Correlation rules:
1. Aggregate summary statistics across all regions to find correlation of the tables 
2. Only compare summary stats from the same level
3. Any statistic from level 1 can be correlated to any other level 1 statistic
4. Average correlation coefficient of all level 1 statistics gives us table correlations
