```python
# This is to just be used whilst drafting the project. I.e before deploying the application. Ideally, you would want to just define a config.py (or any arbitrary format) file that defines the dictionary which could be done by just by saving the self.id2name, self.name2id

class SearchData:
    def __init__(self, meta: pd.DataFrame = None) -> None:
        if meta is None:
            self.meta = pd.read_excel("../data/2021 Census GCP State Electroral Division for NSW/Metadata/Metadata_2021_GCP_DataPack_R1_R2.xlsx", header=10, sheet_name="Cell Descriptors Information")
        else:
            self.meta = meta
            
        self.id2name = dict(self.meta[["Short", "Long"]].values)
        # what u need
        self.name2id = dict(self.meta[["Long", "Short"]].values)
        self.location = dict(self.meta[["Long", "Profiletable"]].values)

    def fetch(self, user_input: str) -> str:
        inputs = user_input.replace(" ", "_")
        print("Column representative: ")
        yield self.name2id[inputs] # gives the column representative
        print("File containing the column: ")
        yield self.location[inputs] # gives the location of the file where the column is located


class UserInput(SearchData):
    def __init__(self):
        super().__init__()
    
    def __call__(self, user_input: str):
        feats = self.fetch(user_input)
        while True:
            try:
                print(next(feats))
            except StopIteration:
                return
    
```
